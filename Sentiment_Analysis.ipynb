**Importing Libraries**

```python
import pandas as pd
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt
```


## Sentiment Analysis Process with TextBlob

For sentiment analysis, we employed TextBlob, a powerful Python library. This process aimed to determine the sentiment—whether a tweet is Positive, Negative, or Neutral.
1. **TextBlob Integration:**
   We integrated TextBlob into our analysis pipeline. TextBlob provides a straightforward API for processing textuallyanalysis_notebook).


```python
#Importing dataset
df = pd.read_csv("cleaned_dataset.csv")



# Function to get sentiment polarity from a text using TextBlob
def get_sentiment(text):
    analysis = TextBlob(str(text))
    # Classify the polarity as positive, negative, or neutral
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment analysis function to the 'tweet' column
df['sentiment'] = df['tweet'].apply(get_sentiment)



# Save the DataFrame with sentiment analysis results to a new CSV file
df.to_csv('tweet_sentiment.csv', index=False)
```

### Word Cloud of Most Frequent Words in Tweets
